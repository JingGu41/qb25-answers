```python

import numpy 
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm
import math

genome_size = 1000000
coverage_3x = 3
readlength = 100

num_reads_3x = int(genome_size * coverage_3x / readlength)

#calculate_number_of_reads(genomesize, readlength, coverage)
## use an array to keep track of the coverage at each position in the genome
genome_coverage_3x = numpy.zeros(genome_size, dtype=int)

print(genome_size)

rng_3x = numpy.random.default_rng()
for i in range(0, num_reads_3x):

  startpos_3x = rng_3x.integers(0,genome_size-readlength+1)
  endpos_3x = startpos_3x + readlength
  genome_coverage_3x[startpos_3x:endpos_3x] += 1

numpy.savetxt("genome_coverage_3x.txt", genome_coverage_3x)

## get the range of coverages observed
maxcoverage_3x = int(genome_coverage_3x.max())
xs_3x = numpy.arange(0, maxcoverage_3x + 1)

## Get the poisson pmf at each of these
#code get from isabel 
poisson_estimates_3x = poisson.pmf(xs_3x, mu=coverage_3x)


## Get normal pdf at each of these (i.e. the density between each adjacent pair of points)
#code get from isabel 
normal_estimates_3x = norm.pdf(xs_3x, loc=coverage_3x, scale=numpy.sqrt(coverage_3x))
 
numpy.savetxt("coverage_values_3x.txt", xs_3x)
numpy.savetxt("poisson_estimates_3x.txt", poisson_estimates_3x)
numpy.savetxt("normal_estimates_3x.txt", normal_estimates_3x)

## now plot the histogram and probability distributions
```

```python
reads = ['ATTCA', 'ATTGA', 'CATTG', 'CTTAT', 'GATTG', 'TATTT', 'TCATT', 'TCTTA', 'TGATT', 'TTATT', 'TTCAT', 'TTCTT', 'TTGAT']
k = 3

graph = set()

for read in reads:
    for i in range(len(read) - k):
        kmer1 = read[i:i+k]
        kmer2 = read[i+1:i+1+k]
        edge = f"{kmer1} -> {kmer2}"
        graph.add(edge)

with open("debruijn_graph.dot", "w") as f:
    f.write("digraph G {\n")
    for edge in sorted(graph):
        f.write(f"    {edge};\n")
    f.write("}\n")
```
